```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
```

```python
data = pd.read_excel("C:/Users/Madhujson/Desktop/student-performance-prediction/student_data.csv.xlsx")
data.head()
```

```python
data.head()
data.info()
```

```python
print(data.columns)
```

```python
X = data[['study hours', 'attendence', 'previous marks']]
y = data['final marks']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

print("Model trained successfully ✅")
```

```python
import pandas as pd

new_data = pd.DataFrame([[5, 78, 72]], 
                        columns=['study hours', 'attendence', 'previous marks'])

prediction = model.predict(new_data)
print("Predicted Marks:", prediction[0])
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)
score = r2_score(y_test, y_pred)

print("Model Accuracy:", score)
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['study hours'], data['final marks'])
plt.xlabel("Study Hours")
plt.ylabel("Final Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
import pickle

pickle.dump(model, open("model.pkl", "wb"))
```
